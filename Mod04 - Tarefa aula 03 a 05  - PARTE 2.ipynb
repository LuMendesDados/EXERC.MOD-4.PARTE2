## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)

df_mat = pd.DataFrame(np.random.randint(24, size=(101, 1)), columns=['Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(101, 1)), columns=['Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(101, 1)), columns=['Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(101, 1)), columns=['Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(101, 1)), columns=['Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(101, 1)), columns=['Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(101, 1)), columns=['Qt_acertos'])

````

````python
# 1) Seu código aqui

alunos = []

frame = pd.DataFrame(alunos, columns=['mat', 'por', 'geo', 'ing', 'his', 'fis', 'qui'], index=range(1,101))

# preenchendo a tabela

frame['mat'] = df_mat
frame['por'] = df_por
frame['geo'] = df_geo
frame['ing'] = df_ing
frame['his'] = df_his
frame['fis'] = df_fis
frame['qui'] = df_qui

# percentual de acerto por disciplina

frame['perc_mat'] = round(df_mat / 24, 2)
frame['perc_por'] = round(df_por / 18, 2)
frame['perc_geo'] = round(df_geo / 8, 2)
frame['perc_ing'] = round(df_ing / 8, 2)
frame['perc_his'] = round(df_his / 8, 2)
frame['perc_fis'] = round(df_fis / 12, 2)
frame['perc_qui'] = round(df_qui / 12, 2)

frame.head()                                                                                                         

````

````python
# 2) total de acertos da prova

frame['total_acertos'] = df_mat + df_por + df_geo + df_ing + df_his + df_fis + df_qui

frame.head()
````

````python
# 3) percentual da prova => nota máxima = 90

frame['perc_prova'] = round(frame['total_acertos'] / 90, 2)

frame.head()
````

````python
# 4) Nota de corte = 45 => quantos alunos foram para 2º fase

frame['fase_2'] = frame['total_acertos'] >= 45

a = frame['fase_2'].sum()

print('Total de alunos na 2ª Fase é', a)

frame.head()     
````
